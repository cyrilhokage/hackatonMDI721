```python
import pandas as pd

df = (pd.read_excel("data/bp_review_2018.xlsx",sheet_name = "Primary Energy Consumption",skiprows=2)
      .dropna(axis=0,subset=[1965]))

df_by_fuel = (pd.read_excel("data/bp_review_2018.xlsx",sheet_name = "Primary Energy - Cons by fuel",skiprows=2))
#       .dropna(axis=0,subset=[1965]))
```

```python
df_primary = (df[~df["Million tonnes oil equivalent"].str.contains("Total|OECD|Asia|European")].iloc[:,:-5]
              .rename(columns ={"Million tonnes oil equivalent":"Country"})
              .set_index("Country"))
```

```python
df_2 = (pd.read_excel("data/population.xls",skiprows=3)).iloc[:,:-1].drop(["Indicator Name","Indicator Code"],axis=1)
mapping = (pd.read_excel("data/population.xls",skiprows=0,sheet_name = 1))[["Country Code","Region"]]
mapping
df_2

merged = pd.merge(df_2,mapping, left_on="Country Code",right_on="Country Code", how='left')

pop_reg = merged.groupby("Region").agg('sum').transpose()
np.array(pop_reg)

# pop_reg.to_csv("population_region")
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# plt.figure(figsize=(25,10))
# plt.legend(pop_reg.columns)
# plt.xticks(rotation=90)
pop_reg.iloc[:,:].plot(figsize=(20,10))
```

```python
pop_reg.plot(kind='bar',stacked = True,figsize = (20,10))
```

```python
from sklearn import linear_model
pop_tot = pop_reg.sum(axis=1)
lm_pop = linear_model.LinearRegression()
lm_pop.fit(np.array(pop_tot.index).reshape(-1,1),pop_tot)
```

```python
pop_pred_2050 = lm_pop.predict(np.array(np.arange(2019,2051)).reshape(-1,1))
pred_years = np.arange(2019,2051)

plt.plot(np.array(pop_tot.index).astype('int'),np.array(pop_tot))
plt.plot(pred_years,pop_pred_2050,color='green',linestyle = '--')
plt.plot(pop_tot)
plt.xlim(1960,2060)

years = np.hstack((np.array(pop_tot.index).astype('int'),pred_years))
pop = np.hstack((np.array(pop_tot),pop_pred_2050))
```

```python
en_cons_tot = df_primary.sum(axis=0)
# en_cons_tot
# pop_tot
```

```python
df_gdp_per_cap = pd.read_csv("data/dsp_per_cap.csv").set_index("Region")
X_gdp = np.array(df_gdp_per_cap.T.index).reshape(-1, 1)
Y_gdp = df_gdp_per_cap.T.mean(axis=1)
test_years = np.arange(2019, 2051).reshape(-1, 1)

reg_gdp = linear_model.LinearRegression().fit(X_gdp, Y_gdp)
predict_gdp = reg_gdp.predict(test_years)

# years_2 = np.hstack((np.array(pop_tot.index).astype('int'),pred_years))
gdp = np.hstack((np.array(Y_gdp),predict_gdp))
gdp


# pred_years.reshape(-1,1)
# predict_gdp.reshape(-1,1)
# pop.reshape(-1,1)
var_regs = np.hstack((years.reshape(-1,1),gdp.reshape(-1,1)))
var_regs = np.hstack((var_regs,pop.reshape(-1,1)))

model = pd.DataFrame(var_regs,columns=["year","GDP_capita","population"]).loc[5:58,:]
model_95_18 = pd.merge(model,en_cons_tot,left_on='year',right_index=True)

X_to_predict = pd.DataFrame(var_regs,columns=["year","GDP_capita","population"]).loc[59:,:]
X_to_predict
# model
# np.concatenate(np.array(Y_gdp),predict_gdp.reshape(-1,))
# predict_gdp.reshape(-1,)
```

```python
X_matrix = model_95_18.iloc[:,:-1]
Y_matrix = model_95_18.iloc[:,-1]

years1 = np.arange(1965,2019)
years2 = np.arange(2019,2051)

linear_forecast_fit = linear_model.LinearRegression()
linear_forecast_fit.fit(X_matrix,Y_matrix)
energy_cons_pred = linear_forecast_fit.predict(X_to_predict)+1000
```

```python
plt.plot(years1,Y_matrix)
plt.plot(years2,energy_cons_pred,color='green',linestyle = '--')
plt.title("World Energy Consuption = f(GDP per capita, Population)")

# plt.xlim(1960,2060)
energy_cons_pred
```
